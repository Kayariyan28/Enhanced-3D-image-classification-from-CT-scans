```python
import os
import zipfile
import numpy as np
import tensorflow as tf
import keras
from keras import layers
import nibabel as nib
from scipy import ndimage
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_auc_score, precision_score, recall_score, f1_score

# Setup
# Ensure TensorFlow uses GPU if available
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        print(e)

# Downloading the MosMedData
# --- FIX APPLIED HERE ---
# Download url of normal CT scans.
url_ct0 = "https://github.com/hasibzunair/3D-image-classification-tutorial/releases/download/v0.2/CT-0.zip"
# fname is just the filename, cache_dir is the path
ct0_fname = "CT-0.zip"
cache_dir_path = os.getcwd()
filename_ct0 = keras.utils.get_file(fname=ct0_fname, origin=url_ct0, cache_dir=cache_dir_path, extract=False) # extract=False because we'll unzip manually

# Download url of abnormal CT scans.
url_ct23 = "https://github.com/hasibzunair/3D-image-classification-tutorial/releases/download/v0.2/CT-23.zip"
ct23_fname = "CT-23.zip"
filename_ct23 = keras.utils.get_file(fname=ct23_fname, origin=url_ct23, cache_dir=cache_dir_path, extract=False) # extract=False

# --- END OF FIX ---

# Make a directory to store the data.
data_dir = os.path.join(os.getcwd(), "MosMedData") # Define data_dir consistently
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

# Unzip data
print(f"Unzipping {filename_ct0} to {data_dir}")
with zipfile.ZipFile(filename_ct0, "r") as z_fp:
    z_fp.extractall(data_dir) # Use data_dir

print(f"Unzipping {filename_ct23} to {data_dir}")
with zipfile.ZipFile(filename_ct23, "r") as z_fp:
    z_fp.extractall(data_dir) # Use data_dir


# --- Enhanced Data Loading and Preprocessing ---
def read_nifti_file(filepath):
    """Read and load volume"""
    scan = nib.load(filepath)
    scan = scan.get_fdata()
    return scan

def normalize(volume):
    """Normalize the volume"""
    min_hu = -1000
    max_hu = 400
    volume[volume < min_hu] = min_hu
    volume[volume > max_hu] = max_hu
    volume = (volume - min_hu) / (max_hu - min_hu)
    volume = volume.astype("float32")
    return volume

def resize_volume(img, desired_depth=64, desired_width=128, desired_height=128, interpolation_order=1):
    """Resize across z-axis and an x-y plane"""
    current_depth = img.shape[-1]
    current_width = img.shape[0]
    current_height = img.shape[1]

    depth_factor = desired_depth / current_depth
    width_factor = desired_width / current_width
    height_factor = desired_height / current_height

    # Rotate (specific to this dataset's orientation needs)
    img = ndimage.rotate(img, 90, reshape=False)
    # Resize
    img = ndimage.zoom(img, (width_factor, height_factor, depth_factor), order=interpolation_order)
    return img

def process_scan(path, desired_depth=64, desired_width=128, desired_height=128):
    """Read, normalize, and resize volume"""
    volume = read_nifti_file(path)
    volume = normalize(volume)
    volume = resize_volume(volume, desired_depth, desired_width, desired_height)
    return volume

# Load scan paths
# --- Ensure paths point to the correct unzipped location ---
normal_scan_paths = [
    os.path.join(data_dir, "CT-0", x) # Use data_dir
    for x in os.listdir(os.path.join(data_dir, "CT-0")) if x.endswith((".nii.gz", ".nii"))
]
abnormal_scan_paths = [
    os.path.join(data_dir, "CT-23", x) # Use data_dir
    for x in os.listdir(os.path.join(data_dir, "CT-23")) if x.endswith((".nii.gz", ".nii"))
]
print(f"CT scans with normal lung tissue: {len(normal_scan_paths)}")
print(f"CT scans with abnormal lung tissue: {len(abnormal_scan_paths)}")

# Prepare data and labels (for initial loading, will be split later in K-Fold)
# Check if paths are valid before processing
if not normal_scan_paths:
    print("Warning: No normal scans found. Check the unzipping process and path.")
if not abnormal_scan_paths:
    print("Warning: No abnormal scans found. Check the unzipping process and path.")

abnormal_scans = np.array([process_scan(path) for path in abnormal_scan_paths])
normal_scans = np.array([process_scan(path) for path in normal_scan_paths])

abnormal_labels = np.array([1 for _ in range(len(abnormal_scans))])
normal_labels = np.array([0 for _ in range(len(normal_scans))])

# Combine all data for K-Fold splitting
all_scans = np.concatenate((abnormal_scans, normal_scans), axis=0)
all_labels = np.concatenate((abnormal_labels, normal_labels), axis=0)
print(f"Total number of samples: {all_scans.shape[0]}")


# --- Enhanced Data Augmentation ---
def random_rotate(volume):
    """Rotate the volume by a few degrees"""
    def scipy_rotate(volume_np):
        angles = [-20, -15, -10, -5, 0, 5, 10, 15, 20]
        angle = random.choice(angles)
        volume_np = ndimage.rotate(volume_np, angle, reshape=False, cval=0) # Use cval to fill with background
        volume_np[volume_np < 0] = 0
        volume_np[volume_np > 1] = 1
        return volume_np.astype(np.float32)

    augmented_volume = tf.numpy_function(scipy_rotate, [volume], tf.float32)
    return augmented_volume

def random_zoom(volume, min_zoom=0.9, max_zoom=1.1):
    """Apply random zoom to the volume using scipy within tf.numpy_function."""
    def scipy_zoom(volume_np):
        current_zoom_factor = np.random.uniform(min_zoom, max_zoom)
        # Ensure zoom factor is applied to each dimension for isotropic zoom
        zoom_factors = [current_zoom_factor] * (volume_np.ndim -1) + [1.0] # Keep depth zoom at 1 or also random
        # If you want depth to also zoom:
        # zoom_factors = [current_zoom_factor] * volume_np.ndim

        zoomed_volume = ndimage.zoom(volume_np, zoom_factors, order=1, mode='constant', cval=0.0)

        # Crop or pad back to original size
        original_shape_np = volume_np.shape
        output = np.zeros_like(volume_np)

        curr_shape = zoomed_volume.shape

        crop_slices = []
        pad_slices = []

        for i in range(volume_np.ndim):
            if curr_shape[i] > original_shape_np[i]: # Crop
                start = (curr_shape[i] - original_shape_np[i]) // 2
                crop_slices.append(slice(start, start + original_shape_np[i]))
                pad_slices.append(slice(None))
            else: # Pad
                crop_slices.append(slice(None))
                pad_start = (original_shape_np[i] - curr_shape[i]) // 2
                pad_slices.append(slice(pad_start, pad_start + curr_shape[i]))

        output[tuple(pad_slices)] = zoomed_volume[tuple(crop_slices)]

        output[output < 0] = 0
        output[output > 1] = 1
        return output.astype(np.float32)

    zoomed_volume = tf.numpy_function(scipy_zoom, [volume], tf.float32)
    zoomed_volume.set_shape(volume.get_shape())
    return zoomed_volume


def random_shift(volume, max_shift_fraction=0.08): # Reduced shift fraction slightly
    """Apply random shift to the volume"""
    input_shape_tf = tf.shape(volume) # H, W, D

    # Use tf.cast on input_shape_tf directly
    max_dx = tf.cast(tf.cast(input_shape_tf[0], tf.float32) * max_shift_fraction, tf.int32)
    max_dy = tf.cast(tf.cast(input_shape_tf[1], tf.float32) * max_shift_fraction, tf.int32)
    max_dz = tf.cast(tf.cast(input_shape_tf[2], tf.float32) * max_shift_fraction, tf.int32)

    dx = tf.random.uniform(shape=[], minval=-max_dx, maxval=max_dx + 1, dtype=tf.int32)
    dy = tf.random.uniform(shape=[], minval=-max_dy, maxval=max_dy + 1, dtype=tf.int32)
    dz = tf.random.uniform(shape=[], minval=-max_dz, maxval=max_dz + 1, dtype=tf.int32)

    volume_shifted = tf.roll(volume, shift=[dx, dy, dz], axis=[0, 1, 2])
    return volume_shifted


def train_preprocessing(volume, label):
    """Process training data by augmenting and adding a channel."""
    volume = random_rotate(volume)
    volume = random_zoom(volume) # Enable if zoom implementation is robust
    volume = random_shift(volume)
    volume = tf.expand_dims(volume, axis=3)
    # Set static shape information
    volume.set_shape([128, 128, 64, 1])
    return volume, label

def validation_preprocessing(volume, label):
    """Process validation data by only adding a channel."""
    volume = tf.expand_dims(volume, axis=3)
    # Set static shape information
    volume.set_shape([128, 128, 64, 1])
    return volume, label


# --- Enhanced Model Definition (ResNet-inspired) ---
def res_block_3d(inputs_tensor, filters, kernel_size=3, strides=1, use_batchnorm=True):
    """A 3D residual block."""
    x = layers.Conv3D(filters, kernel_size, strides=strides, padding="same", activation=None)(inputs_tensor)
    if use_batchnorm:
        x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.Conv3D(filters, kernel_size, strides=1, padding="same", activation=None)(x)
    if use_batchnorm:
        x = layers.BatchNormalization()(x)

    # Shortcut connection
    if strides > 1 or inputs_tensor.shape[-1] != filters:
        shortcut = layers.Conv3D(filters, 1, strides=strides, padding="same")(inputs_tensor)
        if use_batchnorm:
             shortcut = layers.BatchNormalization()(shortcut)
    else:
        shortcut = inputs_tensor

    x = layers.Add()([x, shortcut])
    x = layers.Activation("relu")(x)
    return x

def get_model(width=128, height=128, depth=64, name="3dcnn_resnet_style"):
    """Build a 3D CNN model with residual blocks."""
    inputs = keras.Input((width, height, depth, 1))

    x = layers.Conv3D(filters=32, kernel_size=5, strides=1, padding="same", activation=None)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.MaxPool3D(pool_size=2)(x)


    x = res_block_3d(x, filters=64, strides=1)
    x = layers.MaxPool3D(pool_size=2)(x)

    x = res_block_3d(x, filters=128, strides=1)
    x = layers.MaxPool3D(pool_size=2)(x)

    x = res_block_3d(x, filters=256, strides=1) # Changed stride to 1, maxpool will downsample
    x = layers.MaxPool3D(pool_size=2)(x)


    x = layers.GlobalAveragePooling3D()(x)
    x = layers.Dense(units=512, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(units=1, activation="sigmoid")(x)

    model = keras.Model(inputs, outputs, name=name)
    return model

# --- Training with Cross-Validation ---
N_SPLITS = 3  # Reduced for faster demonstration, recommend 5 for actual use
BATCH_SIZE = 4
EPOCHS = 20 # Reduced for faster demonstration, recommend 50-100
INITIAL_LEARNING_RATE = 0.0001

skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

fold_accuracies, fold_losses, fold_precisions, fold_recalls, fold_aucs = [], [], [], [], []
all_histories = []
all_true_labels_cv, all_pred_probs_cv = [], []


for fold_num, (train_index, val_index) in enumerate(skf.split(all_scans, all_labels)):
    print(f"\n--- Starting Fold {fold_num + 1}/{N_SPLITS} ---")

    x_train_fold, x_val_fold = all_scans[train_index], all_scans[val_index]
    y_train_fold, y_val_fold = all_labels[train_index], all_labels[val_index]

    train_loader_fold = tf.data.Dataset.from_tensor_slices((x_train_fold, y_train_fold))
    validation_loader_fold = tf.data.Dataset.from_tensor_slices((x_val_fold, y_val_fold))

    train_dataset_fold = (
        train_loader_fold.shuffle(len(x_train_fold), reshuffle_each_iteration=True) # Reshuffle
        .map(train_preprocessing, num_parallel_calls=tf.data.AUTOTUNE)
        .batch(BATCH_SIZE)
        .prefetch(tf.data.AUTOTUNE)
    )
    validation_dataset_fold = (
        validation_loader_fold.map(validation_preprocessing, num_parallel_calls=tf.data.AUTOTUNE) # No shuffle for val
        .batch(BATCH_SIZE)
        .prefetch(tf.data.AUTOTUNE)
    )

    model_fold = get_model(width=128, height=128, depth=64, name=f"model_fold_{fold_num+1}")

    lr_schedule = keras.optimizers.schedules.ExponentialDecay(
        INITIAL_LEARNING_RATE, decay_steps=len(x_train_fold) // BATCH_SIZE * 5, # Decay every 5 epochs
        decay_rate=0.96, staircase=True
    )
    optimizer = keras.optimizers.Adam(learning_rate=lr_schedule)

    model_fold.compile(
        loss="binary_crossentropy",
        optimizer=optimizer,
        metrics=[
            "accuracy",
            keras.metrics.Precision(name="precision"),
            keras.metrics.Recall(name="recall"),
            keras.metrics.AUC(name="auc", curve='ROC') # Specify curve for AUC
        ]
    )
    model_fold.summary() # Print summary for one fold to check layers

    checkpoint_filepath = f"3d_image_classification_fold_{fold_num + 1}.weights.h5" # Keras 3 uses .weights.h5
    checkpoint_cb = keras.callbacks.ModelCheckpoint(
        checkpoint_filepath, save_best_only=True, monitor="val_auc", mode="max", save_weights_only=True
    )
    early_stopping_cb = keras.callbacks.EarlyStopping(monitor="val_auc", patience=10, mode="max", restore_best_weights=True) # Reduced patience
    # Removed TensorBoard for brevity in this correction, can be re-added

    print(f"Training model for fold {fold_num + 1}...")
    history = model_fold.fit(
        train_dataset_fold,
        validation_data=validation_dataset_fold,
        epochs=EPOCHS,
        callbacks=[checkpoint_cb, early_stopping_cb],
        verbose=1 # Changed to 1 for more output
    )
    all_histories.append(history)

    model_fold.load_weights(checkpoint_filepath) # Load best weights

    results = model_fold.evaluate(validation_dataset_fold, verbose=0)
    fold_losses.append(results[0])
    fold_accuracies.append(results[1])
    fold_precisions.append(results[2])
    fold_recalls.append(results[3])
    fold_aucs.append(results[4])

    print(f"Fold {fold_num + 1} Val - Loss: {results[0]:.4f}, Acc: {results[1]:.4f}, Prec: {results[2]:.4f}, Rec: {results[3]:.4f}, AUC: {results[4]:.4f}")

    y_pred_prob_fold = model_fold.predict(validation_dataset_fold)
    all_true_labels_cv.extend(y_val_fold)
    all_pred_probs_cv.extend(y_pred_prob_fold.flatten())


# --- Overall Performance Evaluation ---
print("\n--- Cross-Validation Results ---")
if fold_losses: # Check if list is not empty
    print(f"Avg Val Loss: {np.mean(fold_losses):.4f} (Std: {np.std(fold_losses):.4f})")
    print(f"Avg Val Acc: {np.mean(fold_accuracies):.4f} (Std: {np.std(fold_accuracies):.4f})")
    print(f"Avg Val Prec: {np.mean(fold_precisions):.4f} (Std: {np.std(fold_precisions):.4f})")
    print(f"Avg Val Rec: {np.mean(fold_recalls):.4f} (Std: {np.std(fold_recalls):.4f})")
    print(f"Avg Val AUC: {np.mean(fold_aucs):.4f} (Std: {np.std(fold_aucs):.4f})")

    last_history = all_histories[-1].history
    fig, ax = plt.subplots(2, 2, figsize=(15, 10)) # Adjusted for 4 main metrics
    metrics_to_plot = ["accuracy", "loss", "precision", "recall"]
    plot_labels = ["Accuracy", "Loss", "Precision", "Recall"]

    for i, metric in enumerate(metrics_to_plot):
        row, col = i // 2, i % 2
        ax[row, col].plot(last_history.get(metric, []))
        ax[row, col].plot(last_history.get("val_" + metric, []))
        ax[row, col].set_title(f"Model {plot_labels[i]} (Last Fold)")
        ax[row, col].set_xlabel("Epochs")
        ax[row, col].set_ylabel(plot_labels[i])
        ax[row, col].legend(["Train", "Validation"])
    plt.tight_layout()
    plt.show()

    if 'auc' in last_history and 'val_auc' in last_history:
        plt.figure(figsize=(7, 5))
        plt.plot(last_history["auc"])
        plt.plot(last_history["val_auc"])
        plt.title("Model AUC (Last Fold)")
        plt.xlabel("Epochs")
        plt.ylabel("AUC")
        plt.legend(["Train", "Validation"])
        plt.tight_layout()
        plt.show()


    all_pred_labels_cv = (np.array(all_pred_probs_cv) > 0.5).astype(int)
    cm = confusion_matrix(all_true_labels_cv, all_pred_labels_cv)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Normal (0)", "Abnormal (1)"])
    disp.plot(cmap=plt.cm.Blues)
    plt.title("Overall Confusion Matrix (All Folds)")
    plt.show()

    overall_f1 = f1_score(all_true_labels_cv, all_pred_labels_cv)
    print(f"Overall F1-Score (All Folds): {overall_f1:.4f}")


# --- Model Interpretability: 3D Grad-CAM ---
def make_gradcam_heatmap_3d(img_array_gc, model_gc, last_conv_layer_name_gc, pred_index_gc=None):
    grad_model_gc = keras.Model(
        model_gc.inputs, [model_gc.get_layer(last_conv_layer_name_gc).output, model_gc.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output_gc, preds_gc = grad_model_gc(img_array_gc)
        if pred_index_gc is None: # For binary classification, use the single output neuron
            class_channel_gc = preds_gc[:, 0]
        else: # For multi-class
            class_channel_gc = preds_gc[:, pred_index_gc]


    grads_gc = tape.gradient(class_channel_gc, last_conv_layer_output_gc)
    if grads_gc is None:
        print(f"Error: Gradients are None for layer {last_conv_layer_name_gc}. Check layer name and model connectivity.")
        return None

    pooled_grads_gc = tf.reduce_mean(grads_gc, axis=(0, 1, 2, 3))
    last_conv_layer_output_gc = last_conv_layer_output_gc[0]
    heatmap_gc = last_conv_layer_output_gc @ pooled_grads_gc[..., tf.newaxis]
    heatmap_gc = tf.squeeze(heatmap_gc)
    heatmap_gc = tf.maximum(heatmap_gc, 0) / (tf.math.reduce_max(heatmap_gc) + tf.keras.backend.epsilon())
    return heatmap_gc.numpy()


# Use the model from the last fold (model_fold)
grad_cam_model_viz = model_fold
last_conv_layer_name_viz = None
# Find a suitable conv layer before global pooling, often the last one in a res_block or a main conv block
for layer in reversed(grad_cam_model_viz.layers):
    if isinstance(layer, layers.Conv3D) and "relu" in layer.name.lower() or "conv3d" in layer.name.lower() : # More robust check
        # Avoid picking very early layers if possible, aim for deeper ones
        if layer.output.shape[1] > 4 and layer.output.shape[2] > 4 and layer.output.shape[3] > 2: # Heuristic for feature map size
             last_conv_layer_name_viz = layer.name
             break
if last_conv_layer_name_viz is None and len(grad_cam_model_viz.layers) > 5: # Fallback if heuristic fails
    # Try to get one of the last conv layers more directly
    conv_layers = [l for l in grad_cam_model_viz.layers if isinstance(l, layers.Conv3D)]
    if conv_layers: last_conv_layer_name_viz = conv_layers[-1].name


print(f"Attempting Grad-CAM with layer: {last_conv_layer_name_viz}")

if last_conv_layer_name_viz and len(x_val_fold) > 0:
    scan_to_visualize_idx = 0
    single_scan_processed_viz = x_val_fold[scan_to_visualize_idx]
    img_array_for_gradcam_viz = np.expand_dims(np.expand_dims(single_scan_processed_viz, axis=0), axis=-1)

    heatmap_3d_viz = make_gradcam_heatmap_3d(img_array_for_gradcam_viz, grad_cam_model_viz, last_conv_layer_name_viz)

    if heatmap_3d_viz is not None:
        pred_prob_gradcam_viz = grad_cam_model_viz.predict(img_array_for_gradcam_viz)[0][0]
        pred_label_gradcam_viz = "Abnormal" if pred_prob_gradcam_viz > 0.5 else "Normal"
        actual_label_gradcam_viz = "Abnormal" if y_val_fold[scan_to_visualize_idx] == 1 else "Normal"

        print(f"Grad-CAM for scan {scan_to_visualize_idx}: Actual={actual_label_gradcam_viz}, Predicted={pred_label_gradcam_viz} (Prob: {pred_prob_gradcam_viz:.2f})")

        num_rows_montage, num_cols_montage = 2, 4 # Adjusted montage size
        total_slices_montage = num_rows_montage * num_cols_montage
        fig_montage, ax_montage = plt.subplots(num_rows_montage, num_cols_montage, figsize=(16, 8))
        ax_montage = ax_montage.ravel()

        # Resize heatmap to match image slice dimensions if necessary (e.g., due to pooling)
        # This requires knowing the heatmap's original dimensions and the target slice dimensions.
        # For simplicity, assuming heatmap_3d_viz has same H,W as input slices for now.
        # If not, scipy.ndimage.zoom would be needed for heatmap_3d_viz before slicing.

        heatmap_resized_for_plot = ndimage.zoom(heatmap_3d_viz,
                                                (img_array_for_gradcam_viz.shape[1]/heatmap_3d_viz.shape[0],
                                                 img_array_for_gradcam_viz.shape[2]/heatmap_3d_viz.shape[1],
                                                 img_array_for_gradcam_viz.shape[3]/heatmap_3d_viz.shape[2]),
                                                order=1)


        montage_slice_indices = np.linspace(0, img_array_for_gradcam_viz.shape[3] - 1, total_slices_montage, dtype=int)
        for i, slice_idx in enumerate(montage_slice_indices):
            ax_montage[i].imshow(np.squeeze(img_array_for_gradcam_viz[0, :, :, slice_idx]), cmap="gray")
            ax_montage[i].imshow(heatmap_resized_for_plot[:, :, slice_idx], cmap="jet", alpha=0.5)
            ax_montage[i].set_title(f"Slice {slice_idx}")
            ax_montage[i].axis("off")

        plt.suptitle(f"Grad-CAM Montage - Actual: {actual_label_gradcam_viz}, Pred: {pred_label_gradcam_viz}", fontsize=16)
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()
    else:
        print("Grad-CAM heatmap generation failed.")
else:
    print("Skipping Grad-CAM: No suitable layer found or no validation samples.")

print("\n--- End of Script ---")
```
